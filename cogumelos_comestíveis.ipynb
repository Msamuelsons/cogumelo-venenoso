```python
import pandas as pd
pd.set_option('display.max_columns', 23)
dados = pd.read_csv('/content/mushroom_dataset.csv')
```

```python
dados.head()
```

```python
# one-hot-encoder
colunas_encoder = ['cap-shape',	'cap-surface',	'cap-color',	'bruises',	'odor','gill-attachment',	'gill-spacing',	'gill-size',	'gill-color',	'stalk-shape',	'stalk-root',	'stalk-surface-above-ring',	'stalk-surface-below-ring'	,'stalk-color-above-ring','stalk-color-below-ring',	'veil-type',	'veil-color',	'ring-number',	'ring-type',	'spore-print-color'	,'population'	,'habitat']
```

```python
concatenando = pd.get_dummies(dados, columns=colunas_encoder)
concatenando = concatenando.drop('mushroom', axis=1)
```

```python
dados = dados.drop(columns=colunas_encoder, axis=1)
```

```python
df_encoded = pd.concat([dados, concatenando], axis=1)
```

```python
df_encoded.head()
```

```python
x = df_encoded.drop('mushroom', axis=1)
y = df_encoded['mushroom']
```

```python
x.head()
```

```python
pd.set_option('display.max_rows', 117)
x.isnull().sum() / df_encoded.shape[1]
```

```python
y.value_counts()
```

```python
y = y.replace('EDIBLE', 0)
y = y.replace('POISONOUS', 1)
```

```python
y.head()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

modelo = ExtraTreesClassifier(n_estimators=100, random_state=42)
kfold = StratifiedKFold(n_splits=3, shuffle=True)

extraTree = cross_val_score(modelo, x, y,cv=kfold, n_jobs=-1)

print(extraTree.mean())
```

```python
import matplotlib.pyplot as plt

modelo.fit(x, y)

importancias = modelo.feature_importances_

nomes_features = x.columns

df_importancias = pd.DataFrame({'Feature': nomes_features, 'ImportÃ¢ncia': importancias})

df_importancias = df_importancias.sort_values(by='ImportÃ¢ncia', ascending=False)

plt.figure(figsize=(12, 8))
plt.barh(df_importancias['Feature'][:10], df_importancias['ImportÃ¢ncia'][:10])
plt.xlabel('ImportÃ¢ncia')
plt.ylabel('Feature')
plt.title('ImportÃ¢ncia das Features no Modelo ExtraTreesClassifier')
plt.gca().invert_yaxis()
```

```python
import matplotlib.pyplot as plt

contagem = y.value_counts()

plt.figure(figsize=(8, 6))
contagem.plot(kind='bar', color=['green', 'red'])
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.xticks([0, 1], ['ComestÃ­vel', 'Venenoso'], rotation=0)
plt.title('Contagem de Cogumelos ComestÃ­veis e Venenosos')
plt.show()
```
